```python
from tensorflow.keras.preprocessing.text import one_hot
```

```python
sent = [
    "The glass of milk",
    "The glass of juice",
    "The cup of tea",
    "I am a good boy",
    "I am a good developer",
    "Understand the meaning of words",
    "Your video is very good"
]
```

```python
# Define the vocabulary size
voc_size = 10000
```

```python
#One hot representation of every word
one_hot_repr = [one_hot(words, voc_size) for words in sent]
one_hot_repr
```

```python
from keras.layers import Embedding
from keras.utils import pad_sequences
from keras.models import Sequential
```

```python
import numpy as np
```

```python
#We need to all sentences of fixed length of RNN. So we're doing a pad_seq of "0"
#And padding set to 'pre' means adds it at start.  'post' to add at end.
sent_length = 8
embedded_docs = pad_sequences(one_hot_repr, padding='pre', maxlen=sent_length)
embedded_docs
```

```python
## Feature Representation
dim = 10 #I want 10 features in the word embeddings. Each feature is a dimension
```

```python
model = Sequential()
model.add(Embedding(voc_size, dim))
model.compile('adam', 'mse')
```

```python
model.summary()
```

```python
model.predict(embedded_docs)
```
