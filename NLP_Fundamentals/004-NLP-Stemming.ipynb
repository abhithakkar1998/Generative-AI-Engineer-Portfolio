```python
words = [
    "running", "runner", "runs", "easily", "fairly",
    "studies", "studying", "studied", "connected", "connection",
    "playing", "played", "plays", "happily", "happiness",
    "argues", "arguing", "argument", "flies", "flying"
]
```

```python
#Porter Stemmer
from nltk.stem import PorterStemmer
porter_stemmer = PorterStemmer()
```

```python
for word in words:
    print(f"{word} --> {porter_stemmer.stem(word)}")
```

```python
from nltk.stem import LancasterStemmer
lancaster_stemmer = LancasterStemmer()
for word in words:
    print(f"{word} --> {lancaster_stemmer.stem(word)}")
```

```python
from nltk.stem import SnowballStemmer
snowball_stemmer = SnowballStemmer(language='english')
for word in words:
    print(f"{word} --> {snowball_stemmer.stem(word)}")
```

```python
from nltk.stem import RegexpStemmer
regexp_stemmer = RegexpStemmer('ing$|s$|ed$|ly$|ion$')
for word in words:
    print(f"{word} --> {regexp_stemmer.stem(word)}")
```
