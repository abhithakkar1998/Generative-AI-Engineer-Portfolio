```python
#Streaming

import os
from dotenv import load_dotenv
load_dotenv()

from langchain_openai import AzureChatOpenAI
model = AzureChatOpenAI(
    deployment_name=os.getenv("AZURE_OPENAI_LLM_MODEL"),
    azure_endpoint=os.getenv("AZURE_OPENAI_ENDPOINT"),
    openai_api_key=os.getenv("AZURE_OPENAI_API_KEY"),
    api_version=os.getenv("AZURE_OPENAI_API_VERSION", "2023-05-15"),
)
```

```python
model.invoke("Write me a 200 word paragraph on the importance of AI in healthcare.")
```

```python
text_stream = model.stream("Write me a 200 word paragraph on the importance of AI in healthcare.")
```

```python
for chunk in text_stream:
    print(chunk.text, end='', flush=True)
```

```python
for chunk in model.stream("Write me a 500 word paragraph on the importance of AI in healthcare."):
    print(chunk.text, end='', flush=True)
```

```python
#Batching

responses = model.batch([
    "Why do parrots have colorful feathers?",
    "How do airplanes stay in the air?",
    "What is the similarity between a bird and an airplane?",
])

for response in responses:
    print(response)
```

```python
#Limit parallel calls in batching
responses = model.batch([
        "Why do parrots have colorful feathers?",
        "How do airplanes stay in the air?",
        "What is the similarity between a bird and an airplane?",
        "What is the capital of France?",
        "Who is the president of the United States?",
    ],
    config = {
        'max_concurrency': 2,
    }
)

for response in responses:
    print(response)
```
